```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix



file_path = "/content/drive/My Drive/personality_datasert.csv"
df = pd.read_csv(file_path)
display(df.head())


```

```python
df = df[df['Personality'].isin(['Introvert', 'Extrovert'])].copy()
df['target'] = df['Personality'].map({'Introvert': 0, 'Extrovert': 1})

# 3. Prepare features
features = df.drop(['Personality', 'target'], axis=1)
features_encoded = pd.get_dummies(features, drop_first=True)

# 4. Split
X_train, X_test, y_train, y_test = train_test_split(
    features_encoded, df['target'], test_size=0.2, random_state=42
)

# 5. Train
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# 6. Evaluate
y_pred = model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=['Introvert','Extrovert']))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
